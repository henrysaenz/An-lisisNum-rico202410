# Método de bisección

Se debe dividir los intervalos de búsqueda cada vez a su mitad para obtener e valor en el un valor x de una función f(x) se vuelve cero.

```
tol = 1e-6
max_iter = 1000
f = False
num_iter = 0
a = float(input('Digite el límite inferior (a) del intervalo: '))
b = float(input('Digite el límite superior (b) del intervalo: '))

def funcion(x):
    return 6*x - 1

while f == False and num_iter < max_iter:
    x = (a + b) / 2
    if abs(b - a) < tol:
        print('Se ha alcanzado el número máximo de tolerancia para las iteraciones.')
        f = True
    elif abs(funcion(x)) < tol:
        print(f'El valor en el que la imagen de f(x) se vuelve 0 es: {x}, en un total de {num_iter} iteraciones.')
        f = True
    elif funcion(a) * funcion(x) < 0:
        b = x
    else:
        a = x
    num_iter += 1
```
