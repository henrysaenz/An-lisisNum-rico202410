## Matriz tridiagonal

````
import numpy as np

def tridiagonal(alfa, beta, gamma):
    n = len(beta)
    matriz = np.zeros((n, n))
    np.fill_diagonal(matriz, alfa)
    np.fill_diagonal(matriz[1:], beta)
    np.fill_diagonal(matriz[:, 1:], gamma)
    return matriz

def cholesky_test(matriz):
    L = np.zeros_like(matriz)
    n = len(matriz)
    for i in range(n):
        for j in range(i + 1):
            if j == i:
                suma_cuadrados = np.sum(L[j][:j]**2)
                L[j, j] = np.sqrt(matriz[j, j] - suma_cuadrados)
            else:
                suma_productos = np.sum(L[i][:j] * L[j][:j])
                L[i, j] = (matriz[i, j] - suma_productos) / L[j, j]
    return L

# Parámetros alfa, beta y gamma
alfa = 2
beta = [1, 1, 1, 1]  # Longitud de beta determina el tamaño de la matriz
gamma = 2

# Crear matriz tridiagonal
matriz_tridiagonal = tridiagonal(alfa, beta, gamma)

# Probar la factorización de Cholesky
L = cholesky_test(matriz_tridiagonal)
print("Matriz tridiagonal:")
print(matriz_tridiagonal)
print("Factorización de Cholesky (matriz L):")
print(L)
````

## Implementación con matriz tridiagona (-1,2,-1)

````
import numpy as np

def sor(A, b, omega, x0, tol=1e-6, max_iter=1000):
    """
    Implementación del método SOR (Successive Over-Relaxation).

    Parámetros:
    A: numpy.array
        Matriz de coeficientes del sistema lineal.
    b: numpy.array
        Vector de términos independientes.
    omega: float
        Factor de relajación (0 < omega < 2).
    x0: numpy.array
        Vector inicial de la solución.
    tol: float
        Tolerancia: criterio de convergencia.
    max_iter: int
        Número máximo de iteraciones permitidas.

    Devuelve:
    x: numpy.array
        Vector solución aproximada.
    iterations: int
        Número de iteraciones realizadas.
    """
    n = len(b)
    x = x0.copy()
    iterations = 0
    error = tol + 1
    eigenvalues = np.linalg.eigvals(A)
    rho = max(abs(eigenvalues))
    omega = 2 / (1 + np.sqrt(1 - rho**2))

    while error > tol and iterations < max_iter:
        x_old = x.copy()

        for i in range(n):
            sigma = 0
            for j in range(n):
                if j != i:
                    sigma += A[i, j] * x[j]
            x[i] = (1 - omega) * x[i] + (omega / A[i, i]) * (b[i] - sigma)

        error = np.linalg.norm(x - x_old)
        iterations += 1

    return x, iterations

# Ejemplo de uso
# A = np.array([[4, -1, 0, 0],
#               [-1, 4, -1, 0],
#               [0, -1, 4, -1],
#               [0, 0, -1, 4]])

# b = np.array([5, 5, 0, 5])

x0 = np.zeros_like(b)
omega = 1.2
tol = 1e-6
max_iter = 1000

solucion, num_iteraciones = sor(A, b, omega, x0, 1e-6, 1000)

print("Solución aproximada:", solucion)
print("Número de iteraciones:", num_iteraciones)
````

````
import numpy as np


import numpy as np

# Función para calcular omega óptima
def calculate_optimal_omega(A):
    eigenvalues = np.linalg.eigvals(A)
    rho = max(abs(eigenvalues))
    omega = 2 / (1 + np.sqrt(1 - rho**2))
    return omega

# Luego descomponemos el SOR
def sor_decomposition(A, omega):
    n = A.shape[0]
    D = np.diag(np.diag(A))
    E = np.tril(A, k=-1)
    F = np.triu(A, k=1)
    
    D_minus_wE = D - omega * E
    wF_plus_D_minus_wE = omega * F + (1 - omega) * D
    
    D_minus_wE_inv = np.linalg.inv(D_minus_wE)
    
    return D_minus_wE_inv, wF_plus_D_minus_wE

# Función SOR
def sor(A, b, omega, x0, tol=1e-6, max_iter=1000):
    n = len(b)
    x = x0.copy()
    iterations = 0
    error = tol + 1
    
    # Descomposición SOR de la matriz A
    D_minus_wE_inv, wF_plus_D_minus_wE = sor_decomposition(A, omega)
    
    while error > tol and iterations < max_iter:
        x_old = x.copy()
        
        # Iteración SOR
        x = np.dot(D_minus_wE_inv, np.dot(wF_plus_D_minus_wE, x) + np.dot(D_minus_wE_inv, b))
        
        error = np.linalg.norm(x - x_old)
        iterations += 1
    
    return x, iterations

# Ejemplo de uso
A = np.array([[6, 1, 1, 1, 1],
              [1, 8, 1, 1, 1],
              [1, 1, 10, 1, 1],
              [1, 1, 1, 9, 1],
              [1, 1, 1, 1, 7]])
b = np.array([10, 20, 30, 40, 50])
x0 = np.zeros_like(b)
omega = calculate_optimal_omega(A)  # Calcula el omega óptimo
tol = 1e-6
max_iter = 1000

solucion, num_iteraciones = sor(A, b, omega, x0, tol, max_iter)

print("Solución aproximada:", solucion)
print("Número de iteraciones:", num_iteraciones)
````
