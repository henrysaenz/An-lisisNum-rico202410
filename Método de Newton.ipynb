# Método de Newton

En el método de Newton se selccióna un punto inicial al azar y se traxa la tangente sobre la imagen de se punto.

```
import numpy as np
import matplotlib.pyplot as plt
```

```
def funcion(x):
    return x**3 - 2*x**2 +1
def tangente(x):
    return derivada * (x-x0) + f0
```

### Gráfica inicial de la tangente en un punto dado

```
# f = funcion(3)
tol = 1e-6
x0 = float(input('Digite el punto inicial para la busqueda: '))
f0 = funcion(x0)
derivada = (funcion(x0 + tol) - funcion(x0 - tol))/(2*tol)


#Dado que tomé la ecuación punto pendiente y - y0 = m * (x - x0)
#Rango de valores para la gráfica
x_values = np.linspace(x0 -2, x0 +2, 100)
tangent_values = tangente(x_values)
```

```
#Gráfica de la función y de la tangente
plt.plot(x_values, funcion(x_values), label = 'F(x)')
plt.plot(x_values, tangent_values, label='Tangente en el punto dado')
plt.scatter(x0, f0, color = 'red', label = 'Punto dado')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Tangente en un punto de una función')
plt.legend()
plt.grid(True)
plt.show()
```

### Gráfica de la tangente hasta hallar el punto f(x) = 0

```
def funcion(x):
    return x**3 - 2*x**2 +1

def intercepto(x):
    f0 = funcion(x)
    derivada = (funcion(x + tol) - funcion(x - tol)) / (2 * tol)
    return x - (f0/derivada)

def raiz():
    x0 = float(input('Digite el punto inicial para la búsqueda: '))
    iter_count = 0

    while True:
        xk1 = intercepto(x0)
        iter_count +=1
        print(f"Iteración {iter_count}: x = {xk1}")
        if abs(xk1 - x0) < tol or iter_count >= max_iter:
            break
        x0 = xk1
    return xk1
```

```
tol = 1e-6
max_iter = 100
ra = raiz()
print('La raíz encontrada es: ', ra)
```

```
x_values = np.linspace(-10, 10, 100)
plt.plot(x_values, funcion(x_values), label='Función')
plt.scatter(ra, 0, color='red', label='Raíz')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Encontrar raíz por tangentes sucesivas')
plt.legend()
plt.grid(True)
plt.show()
```

### Gráfica para hallar la ráiz de f(x) trazando todas las tangenes

```
def funcion(x):
    return x**3 - 2*x**2 +1

def intercepto(x):
    f0 = funcion(x)
    derivada = (funcion(x + tol) - funcion(x - tol)) / (2 * tol)
    return x - (f0 / derivada)

def raiz():
    x0 = float(input('Digite el punto inicial para la búsqueda: '))
    tangente_points = [(x0, 0)]

    iter_count = 0
    while True:
        xk1 = intercepto(x0)
        tangente_points.append((xk1, 0))
        iter_count += 1

        # Comprobar la condición de convergencia
        if abs(xk1 - x0) < tol or iter_count >= max_iter:
            break
        
        x0 = xk1

    return tangente_points
```

```
tol = 1e-6
max_iter = 100
tangente_points = raiz()
```

```
x_values = np.linspace(1.5, 3.2, 100)
plt.plot(x_values, funcion(x_values), label='Función')
for i in range(len(tangente_points) - 1):
    plt.plot([tangente_points[i][0], tangente_points[i+1][0]], [tangente_points[i][1], tangente_points[i+1][1]], color='red')
plt.scatter([point[0] for point in tangente_points], [point[1] for point in tangente_points], color='red', label='Tangentes')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Encontrar raíz por tangentes sucesivas')
plt.legend()
plt.grid(True)
plt.show()
```

```
def funcion(x, y):
    return x**2 + y**2 - 4

def intercepto(x, y):
    f0 = funcion(x, y)
    derivada_x = (funcion(x + tol, y) - funcion(x - tol, y)) / (2 * tol)
    derivada_y = (funcion(x, y + tol) - funcion(x, y - tol)) / (2 * tol)
    pendiente = -derivada_x / derivada_y
    return x - (f0/derivada_y)

def raiz():
    x0 = float(input('Digite el punto inicial para la búsqueda: '))
    y0 = float(input('Digite el punto inicial para la búsqueda de y: '))
    iter_count = 0

    while True:
        xk1 = intercepto(x0, y0)
        iter_count +=1
        print(f"Iteración {iter_count}: x = {xk1}")
        if abs(xk1 - x0) < tol or iter_count >= max_iter:
            break
        x0 = xk1
    return xk1
```

```
tol = 1e-6
max_iter = 100
ra = raiz()
print('La raíz encontrada es: ', ra)
```
